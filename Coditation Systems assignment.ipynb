```python
                               #########GAME OF CONNECT FOUR###################

#Complete code is written in Python
    
print('##WELCOME TO THE GAME OF CONNECT FOUR##')
print('---------------------------------------')
row=int(input('Choose the height of the Game:')) #choosing the height and width of game and no. of connections one should get to win a game 
column=int(input('Choose the width of the Game:'))
pieces=int(input('Choose the number of connections to win a Game(enter only positive integer):'))


class connectpieces:
    def __init__(self, row, column,pieces): #initialising class of connectpieces
        self.row = row
        self.column = column
        self.pieces=pieces
        self.board = [['' for x in range(column)] for i in range(row)]  #creating a gameboard  of lenght r*c

    def return_board(self):  #returning that matrix
        return self.board

    def get_column(self, index):         #recognizing the connecting columns in the board 
        return [i[index] for i in self.board]

    def get_row(self, index):       #recognizing the connecting rows in board
        return self.board[index]

    def get_diagonals(self):         #recognizing the connecting diagonals of board
        diagonals = []
        for p in range(self.row + self.column - 1):
            diagonals.append([])
            for q in range(max(p - self.row + 1, 0),
                           min(p + 1, self.row)):
                diagonals[p].append(self.board[self.row - p + q - 1][q])
        for p in range(self.row + self.column - 1):
            diagonals.append([])
            for q in range(max(p - self.row + 1, 0),
                           min(p + 1, self.row)):
                diagonals[p].append(self.board[p - q][q])
        return diagonals

    def make_move(self, team, col):       #playing the game, choosing the block to enter
        if '' not in self.get_column(col):
            return self.board
        i = self.row - 1
        while self.board[i][col] != '':
            i -= 1
        self.board[i][col] = team
        return self.board

    def check_win(self):        #checking the winning condition of the game after every move
        s=self.pieces
        z=[]
        y=[]
        for i in range(s):
            z.append('0')
            y.append('1')
            r=z,y
            l=list(r)    

    
        for i in range(self.row):  # checking rows
            for x in range(self.column - (s-1)):
                if self.get_row(i)[x:x + s] in l:
                    return self.board[i][x]
        for i in range(self.column):  # checking columns
            for x in range(self.row - (s-1)):
                if self.get_column(i)[x:x + s] in l:
                    return self.board[x][i]
                
        for i in self.get_diagonals():  #checking diagonally
            for x in range(len(i)):
                if i[x:x + s] in l:
                    return i[x]

        return None
    
    
b = connectpieces(row,column,pieces)   #calling the class

def Game():   ##GAMING
    while True:
        for i in b.return_board():
            print(i)
        if b.check_win() != None:
            break
        col = int(input('Player 1 choose a column you wish to place your piece: '))-1
        b.make_move('0' , col)
        for i in b.return_board():
            print(i)
        if b.check_win() != None:
            break
        col = int(input('Player 2 choose a column you wish to place your piece: '))-1
        b.make_move('1' , col)
 
    print('##Cheers##\n*You are the Winner*')             


                        
##Opting to play
while True:
    
    choice = int(input("1. New Game\n0. Exit\nEnter your choice here: "))
    if choice==1:
        Game()
    elif choice==0:
        break
    else:
          print("Invalid choice.Please choose an option from below: ")
    

print('Thank you for playing the Game')     
print('-------------------------------')
```
